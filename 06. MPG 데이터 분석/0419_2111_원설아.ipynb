```python
import pandas as pd
import numpy as np

mpg = pd.read_csv('./mpg.csv', encoding='utf-8')
mpg
```

```python
mpg.groupby('category').agg( mean_cty = ('cty', 'mean') ).sort_values('mean_cty', ascending=False)
mpg.groupby('category')['cty'].mean().sort_values(ascending=False)
```

```python
mpg.groupby('manufacturer')['hwy'].mean().sort_values(ascending=False).head(3)
mpg.groupby('manufacturer').agg( mean_hwy = ('hwy', 'mean') ).sort_values('mean_hwy', ascending=False).head(3)
```

```python
mpg[ mpg.category == 'compact'].groupby('manufacturer').agg(n=('manufacturer','count')).sort_values('n', ascending=False)
mpg[ mpg.category == 'compact'].value_counts('manufacturer')
```
