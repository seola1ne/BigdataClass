```python
import pandas as pd
import numpy as np
exam = pd.read_csv('./exam_sample.csv', encoding='utf-8')
exam.head()
```

```python
df1 = exam.groupby(['class'] )
df1
df1.get_group('A')
```

```python
exam.groupby( 'class' ).mean()
df1.mean()
```

```python
exam.groupby( [ 'class', 'sex' ] ).mean()
```

```python
exam['math'].groupby(exam['class']).mean()
exam.groupby('class')['math'].mean()
exam.groupby('class').agg(  mean_math = ( 'math', 'mean' ) )
```

```python
exam.groupby('class').agg( mean_math = ('math','mean'),
                          median_english = ('english', 'median'),
                          n = ('class', 'count') )
```

```python
exam.groupby('class').agg( mean_math = ('math', 'mean') )
```

```python
exam.groupby('class').agg( [ 'mean', 'median', 'count'] )
exam.groupby('class').agg({ 'math':['min','max'] , 'english':'mean'  } )
```
